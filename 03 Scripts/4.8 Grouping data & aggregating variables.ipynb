### Group/aggregate variables

```python
# create a group of 'product_name' values
df.groupby('product_name')
```

```python
# create aggregate mean of 'order_number' based on 'department_id'
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# include 'min' and 'max' to above aggregation
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
# create new 'max_order' column using the transform() command
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head()
```

```python
# apply 'mean' command to entire database, not subset
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

The majority of the columns are within the same basic range of numbers between the subset and the whole dataframe. There are 7 departments that have changed by a point or more when looking at the whole dataframe: 3, 5, 10, 11, 12, 19, 21. It would be interesting to look further into why these are so different when taking the whole dataframe into consideration. 

### Create loyalty_flag

```python
# create a flag to assign loyalty type to customers based on orders
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

### Goup/aggregate with loyalty_flag

```python
# identify spending habits of customers based on loyalty flag
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'max', 'min', 'sum']})
```

Loyal customers appear to spend less per product than the other two customer types. Regular customers have spent the most overall, even above loyal customers. It appears that a loyal customer isn't the highest spending customer on average. 

### Create spending flag

```python
# create flag for different types of spenders: low spender, high spender
ords_prods_merge.loc[ords_prods_merge['prices'] < 10, 'spender_type'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['prices'] >=10, 'spender_type'] = 'High spender'
```

### Create frequency_flag

```python
# create flag for frequent/non-frequent customers using mean of 'days_since_previous_order'
ords_prods_merge['order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_previous_order'].transform(np.median)
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_frequency'] > 10) & (ords_prods_merge['order_frequency'] <=20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] <=10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.shape
```

It appears that there are a lot more frequent customers than non-frequent
