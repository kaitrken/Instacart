### Loyalty bar chart

```python
# create loyalty_flag chart
loyalty_bar = ords_merge['loyalty_flag'].value_counts().plot.bar(rot = 0, color = ['blue', 'gold', 'green'])
```

### Busiest days of week

```python
# create chart showing busiest days of the week
busiest_dow = ords_merge['order_day_of_week'].plot.hist(bins = 7)
```

### Popular products

```python
# create popular department chart
popular_bar = ords_merge['department_id'].value_counts().plot.bar(rot = 0)
```

### Ordering based on loyalty

```python
# check spending habits of regions
loyal_spend_crosstab = pd.crosstab(ords_merge['department_id'], ords_merge['loyalty_flag'], dropna = False)
```

```python
# create bar chart of loyal_spend_crosstab
loyal_spend_bar = loyal_spend_crosstab.plot.bar(rot = 0, color = ['gold', 'green', 'blue'])
```
