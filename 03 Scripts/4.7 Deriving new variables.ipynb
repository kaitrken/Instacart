### Create user-defined function for prices

```python
# creat user-defined function in an if statement
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

### Create loc() if-statement for prices

```python
# use loc() statement to create if-statement
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# apply loc() function to entire dataframe
df_ords_prods.loc[df_ords_prods['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods.loc[(df_ords_prods['prices'] <= 15) & (df_ords_prods['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

### Create for-loop for busiest day

```python
# create a for-loop to find busiest day for orders
df_ords_prods.head()
```

```python
df_ords_prods['order_day_of_week'].value_counts(dropna = False)
```

```python
result = []

for value in df_ords_prods['order_day_of_week']:
  if value == 0:
    result.append('Busiest day')
  elif value == 4:
    result.append('Least busy')
  else:
    result.append('Regularly busy')
```

```python
result
```

```python
df_ords_prods['busiest_day'] = result
```

```python
df_ords_prods['busiest_day'].value_counts(dropna = False)
```

### Create busiest_days column

```python
# change name/results of 'busiest_day' column to 'busiest_days'
result = []

for value in df_ords_prods['order_day_of_week']:
  if value == 0 and 1:
    result.append('Busiest day')
  elif value == 3 and 4:
    result.append('Least busy')
  else:
    result.append('Regularly busy')
```

```python
df_ords_prods['busiest_days'] = result
```

```python
df_ords_prods['busiest_days'].value_counts(dropna = False)
```

The new results column decreased the amount of regularly busy days, remained the same on busiest day, and increased the amount of least busy days

### Create busiest_period_of_day

```python
# create column for 'busiest_period_of_day'
df_ords_prods['order_hour_of_day'].value_counts(dropna = False)
```

the values can be split into 8 different categories to find the most, fewest, and average orders

```python
hours = []

for hour in df_ords_prods['order_hour_of_day']:
  if hour in [10,11,14,15,13,12,16,9]:
    hours.append('Most orders')
  elif hour in [3,4,2,5,1,0,6,23]:
    hours.append('Fewest orders')
  else:
    hours.append('Average orders')
```

```python
df_ords_prods['busiest_period_of_day'] = hours
```
