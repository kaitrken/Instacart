```python
# create subset of sample for 'prices' & 'order_hour_of_day'
df = small[['order_hour_of_day', 'prices']]
```

### Line chart of sampled data

```python
sns.lineplot(data = df, x='order_hour_of_day', y='prices')
```

It appears from the linechart that prices peak in the early morning and then have a very slow downward trend as the day goes on.

### Sampled subset of age, dependents

```python
# create subset of sample for 'age' and 'num_dependents'
fam = small[['age', 'num_dependents']]
```

### Line chart of sampled age data

```python
sns.lineplot(data = fam, x = 'age', y = 'num_dependents')
```

This line chart shows that there isn't a correlation between age and the number of dependents that a customer might have. It varies widely across the spectrum, peaking around age 60 but never really creating a clear pattern

### Age/income scatterplot

```python
# create scatterplot of age and income
sns.scatterplot(x = 'age', y = 'income', data = merge)
```

The majority of customers using instacart have less than a 200000 income. However, as the ages increase the amount of higher income also increases with a higher density of 40+ customers having income up to 300,000. Additionally, no customer under the age of 40 has an income over 400,000.
